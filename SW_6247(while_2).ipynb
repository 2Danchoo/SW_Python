while문을 이용해 아래와 같이 별(*)을 표시하는 프로그램을 만드십시오.

<br><hr/>

입력 : 입력값 없음 <br>
출력 : 
```
*******
 *****
  ***
   *
```

출력값
```
*******　# 별 7개
 *****　# 별 5개, 1
  ***　# 별 3개, 2
   *　# 별 1개, 3
```

## Think Solutions


**1. 별(*) 표시**

1-1. 별의 개수를 **리스트**에 담아 while 문을 사용하여 출력   
1-2. 별의 개수가 규칙적으로 변하므로 (-2) 이를 이용하여 출력


**2. 문자열 정렬**

2-1. 띄어쓰기 칸의 개수를 세어 그 수를 리스트에 담은 후 문자열 형태로 별 표시에 더하여 출력   
2-2. 띄어쓰기 칸의 개수도 규칙적으로 변하므로 (-1) 연산자로 계산하여 출력하는 방법 시도   
2-3. 우선 별(*)만 출력한 뒤 그 문자열을 가운데 정렬하는 함수인 center() 혹은 formatting 함수를 사용하여 출력하는 방법도 가능한지 시도

## Process

1-1. 리스트를 이용하여 별(*) 출력

````
num = [7, 5, 3, 1]
i = 0

while i < len(num) :
    print("*" * num[i])
    i += 1
````

1-2. 규칙을 이용한 별(*) 출력

별 개수 : 7 → 5 → 3 → 1 ( -2 )   
띄어쓰기 칸 : 0 → 2 → 4 → 6 ( +2 )

````
a = 7

while a > 0 :
    print("*" * a)
    a -= 2
````

시도 2. format 메서드를 사용하여 문자열 정렬

```
'{인덱스:<길이}'.format(값)
```

예제
```
[in] '{0:<10}'.format('python')
[out] 'python    '
```

````
num = [7, 5, 3, 1]
i = 0

while i < len(num) :
    print("{:^10} *".format() * num[i])
    i += 1
````

시도 3. 

````
star = [7, 4, 3, 1]

while i < len(star) :
    print("*" * star[i].center(7))  # center 함수는 '문자열' 가운데 정렬 
    i += 1
````

## Python Syntax

### format 메서드로 값을 여러 개 넣는 방법

```
'Hello, {0} {2} {1}'.format('Python', 'Script', 3.6)
'Hello, Python 3.6 Script'
```

{ }에 인덱스를 지정한 뒤 format에는 인덱스가 증가하는 순서대로 값을 넣는다.   
여기서는 {0} {2} {1}처럼 인덱스 순서를 섞었고 format에는 'Python', 'Script', 3.6 순으로 넣었다.    
따라서 문자열에는 인덱스 숫자에 해당하는 값이 들어가서 'Hello, Python 3.6 Script'가 나온다.   

image.png

```
'Hello, {} {} {}'.format('Python', 'Script', 3.6)
'Hello, Python Script 3.6'
```

만약 { }에서 인덱스를 생략하면   
format에 지정한 순서대로 값이 들어간다.

````
i, a, b = 4, 7, 0

while(i):
    print("{}{}{}".format(" "*b, "*"*a, " "*b))
    a-=2; b+=1
    i-=1
````

문자열 정렬

````
star = [7, 5, 3, 1]
num = [0, 1, 2, 3 ]
i = 0

while i < len(star) :
    print(" " * num[i] + "*" * star[i])
    i += 1
````
